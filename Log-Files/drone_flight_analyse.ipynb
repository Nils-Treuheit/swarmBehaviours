```python
%matplotlib widget

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


FILE1 = "logs/ConSteer/ac_50_debug.csv"

df1 = pd.read_csv(FILE1,skiprows=1)
df1.columns = [s.strip() for s in df1.columns]
df1
```

```python
FILE2 = "logs/ConSteer/ac_51_debug.csv"

df2 = pd.read_csv(FILE2,skiprows=1)
df2.columns = [s.strip() for s in df2.columns]
df2
```

```python
fig = plt.figure(figsize=(15,7.5))
ax1 = fig.add_subplot(121,projection='3d')
ax2 = fig.add_subplot(122,projection='3d')
res1 = ax1.scatter(*df1[['pos_x','pos_y','pos_z']].to_numpy().T, c=np.arange(len(df1)))
ax1.set_xlim(min(df1[['pos_x','pos_y']].min(axis=0)),max(df1[['pos_x','pos_y']].max(axis=0)))
ax1.set_ylim(min(df1[['pos_x','pos_y']].min(axis=0)),max(df1[['pos_x','pos_y']].max(axis=0)))
res2 = ax2.scatter(*df2[['pos_x','pos_y','pos_z']].to_numpy().T, c=np.arange(len(df2)))
ax2.set_xlim(min(df2[['pos_x','pos_y']].min(axis=0)),max(df2[['pos_x','pos_y']].max(axis=0)))
ax2.set_ylim(min(df2[['pos_x','pos_y']].min(axis=0)),max(df2[['pos_x','pos_y']].max(axis=0)))
fig.colorbar(res2, ax=ax2);
```

```python
#for c in df1.columns:
#    if "align" in c: print(c)
align_frame1 = df1[[f"{i}_align" for i in range(8)]].astype(float)
align_frame2 = df2[[f"{i}_align" for i in range(8)]].astype(float)
#align_frame1
```

```python
mins1 = align_frame1.min(axis=1,skipna=True)*1.25
maxs1 = align_frame1.max(axis=1,skipna=True)*1.25
mins2 = align_frame2.min(axis=1,skipna=True)*1.25
maxs2 = align_frame2.max(axis=1,skipna=True)*1.25

fig = plt.figure(figsize=(15,7.5))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)
ax1.plot(mins1)
ax1.plot(maxs1)
ax2.plot(mins2)
ax2.plot(maxs2);
```

```python
intrest_frame1 = df1[[s for s in df1.columns if "interest" in s]]
intrest_frame2 = df2[[s for s in df2.columns if "interest" in s]]

mins1 = intrest_frame1.min(axis=1,skipna=True)
maxs1 = intrest_frame1.max(axis=1,skipna=True)
mins2 = intrest_frame2.min(axis=1,skipna=True)
maxs2 = intrest_frame2.max(axis=1,skipna=True)

fig = plt.figure(figsize=(15,7.5))
ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(223)
ax4 = fig.add_subplot(224)
ax1.plot(mins1)
ax1.plot(maxs1)
ax2.plot(mins2)
ax2.plot(maxs2)
ax3.plot(mins1[:800])
ax3.plot(maxs1[:800])
ax4.plot(mins2[:800])
ax4.plot(maxs2[:800]);
```
